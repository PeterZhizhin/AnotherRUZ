```python
import requests
```

```python
auditoriums = requests.get('http://ruz.hse.ru/ruzservice.svc/auditoriums').json()
```

```python
buildings = set(i['building'] for i in auditoriums)
```

```python
buildings_count = {i: sum(1 for j in auditoriums if j['building'] == i) for i in buildings}
```

```python
sorted(buildings_count.items(), key=lambda x: x[1], reverse=True)
```
